```python
import numpy as np
import matplotlib.pyplot as plt

data = [[2, 81], [4, 93], [6, 91], [8, 97]]
x = [i[0] for i in data]
y = [i[1] for i in data]

print(x, y)

plt.figure(figsize=(8, 5))
plt.scatter(x, y)
plt.show()
```

```python
x_data = np.array(x)
y_data = np.array(y)

# ----- 형태의 직선으로 초기화 y=0
a=0
b=0

# 학습률 0.03
# 이거
lr = 0.03

epochs = 2001

for i in range(epochs):
  a_diff = 2/len(x_data)*sum((a*x_data+b-y_data)*x_data)
  b_diff = 2/len(x_data)*sum(a*x_data+b-y_data)
  a=a-lr*a_diff
  b=b-lr*b_diff
  if i % 100 == 0:
    print("epoch=%.f, 기울기=%.04f, 절편=%.04f" % (i, a, b))
```

```python
y_pred = a * x_data + b
plt.scatter(x, y)

plt.plot(x_data, y_pred)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
import platform # 한글 폰트 설정
if platform.system() == 'Windows':# 윈도우인 경우
    font_path = "c:/Windows/Fonts/malgun.ttf" # 맑은 고딕 폰트 경로
    font_name = font_manager.FontProperties(fname=font_path).get_name()
    plt.rc('font', family=font_name)
elif platform.system() == 'Darwin': # macOS인 경우
    plt.rc('font', family='AppleGothic')
else: # Linux 등 다른 OS인 경우
    plt.rc('font', family='NanumGothic')
# 음수 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

data = [[2, 81], [4, 93], [6, 91], [8, 97]]
x = [i[0] for i in data]
y = [i[1] for i in data]

plt.figure(figsize=(8, 5))
plt.scatter(x, y)
plt.show()
```

```python
data = [[1, 0.2], [2, 0.3], [3, 0.5], [4, 0.6], [5, 0.9], [6, 0.95], [7, 1.1], [8, 1.5]]
x = [i[0] for i in data]
y = [i[1] for i in data]

x_data = np.array(x)
y_data = np.array(y)
a = 0
b = 0
lr = 0.03
epochs = 2001

a_hist = []
b_hist = []
a_diff_hist = []
b_diff_hist = []

for i in range(epochs):
  a_diff = 2/len(x_data)*sum((a*x_data+b-y_data)*x_data)
  b_diff = 2/len(x_data)*sum(a*x_data+b-y_data)
  a=a-lr*a_diff
  b=b-lr*b_diff

  a_hist.append(a)
  b_hist.append(b)
  a_diff_hist.append(a_diff)
  b_diff_hist.append(b_diff)

  if i % 100 == 0:
    print('epoch=%.f, 기울기=%.04f, 절편=%.04f' % (i, a, b))

y_pred = a * x_data + b
plt.figure(figsize=(12, 6))

disp_epodh = 100
plt.subplot(1, 2, 1)
plt.plot(a_hist[:disp_epodh], label='a (slope)')
plt.plot(b_hist[:disp_epodh], label='b (intercepter)')
plt.xlabel('Epoch')
plt.ylabel('Value')
plt.title('Gradient Descent for a and b')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(a_diff_hist[:disp_epodh], label='a_diff')
plt.plot(b_diff_hist[:disp_epodh], label='b_diff')
plt.xlabel('Epoch')
plt.ylabel('Gradient')
plt.title('Changes in a_diff and b_diff')
plt.legend()

plt.show()

plt.scatter(x, y)
plt.plot(x_data, y_pred, color='red')
plt.title('Data and Fitted Line')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
